```python
from bs4 import BeautifulSoup
```

```python
html = """
<html>
    <body>
        <p class="intro">This is a paragraph with class "intro".</p>
        <p id="unique-para">This is a unique paragraph with id "unique-para".</p>
        <div class="container">
            <h1 class="header">This is the first h1</h1>
            <h1 class="header">This is the second h1</h1>
            <h1 class="header">This is the third h1</h1>
            <span>This is a span element inside the div.</span>
        </div>
    </body>
</html>
"""
```

```python
soup=BeautifulSoup(html,'html.parser') 
```

```python
print(soup.prettify)
```

```python
para=soup.find('p' , {'class':'intro'})
print(para)
```

```python
header=soup.find_all('h1',{'class':'header'})
print(header)
```

```python
for item in header:
    print(item.text)
    
```

```python
# can use css selector which find_all and find cant 
select_one= soup.select_one('#unique-para')
```

```python
print(select_one)
```

```python
select_fn=soup.select('.header')
print(select_fn)
```

```python
# find : used for single match
# find_all: used for multiple matches and returns list
# select_one :used for single matches using CSS selector
# select: used for multiple matches and returns list using CSS selector
```

```python
import multiprocessing
import requests
import time
import warnings
import os
warnings.filterwarnings('ignore')
 
def download_image(url,index,directory):
    response = requests.get(url, verify=False)
    with open(os.path.join(directory, f'image_{index}.jpg'), 'wb') as f:
        f.write(response.content)
    print(f"Downloaded image_{index}.jpg")
 
def download_images_serially(url, start_index, end_index,directory):
    for i in range(start_index, end_index + 1):
        download_image(url,i,directory)
 
if __name__ == "__main__":
    multiprocessing.freeze_support()
    url = "https://picsum.photos/2000/3000"
    directory="C:\\Python_Training\\Days\\Assignment_5\\downloaded_images"
 
    if not os.path.exists(directory):
        os.makedirs(directory)
   
    print("\nDownloading images 1 to 5 serially...")
    start_time=time.time()
    download_images_serially(url,1,5,directory)
    end_time=time.time()
    time_req=end_time-start_time
    print(f"\nTime required for serial image download is : {round(end_time-start_time,2)} seconds")
 
    print("\nDownloading images 6 to 10 using multiprocessing...")
    processes = []
    start2=time.time()
    for i in range(6, 11):
        p = multiprocessing.Process(target=download_image, args=(url,i,directory))
        processes.append(p)
        p.start()
 
    for p in processes:
        p.join()
    end2=time.time()
    print(f"\nTime required for images download using Multiprocessing is : {round(end2-start2,2)} seconds")
 
    print("All images downloaded.")
```
